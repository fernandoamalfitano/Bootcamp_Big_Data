## 1- Creacion de una matriz con numeros (float) aleatorios entre 0 y 1

```python
def make_constant_matrix(rows: int, cols: int, value: float)->list[list[float]]:
  """
  Esta función recibe el tamaño de la matriz que debe
  de crear (filas y columnas)
  y en cada posición mete un valor constante (value)
  """
  matrix = []
  for i in range(0,cols):
    empty_row = []
    for j in range(0,rows):
      empty_row.append(value)

    matrix.append(empty_row)
  return matrix
```

```python
make_constant_matrix(2,2, 2.4)
```

```python
assert make_constant_matrix(2,2,2.4) == [[2.4,2.4],[2.4,2.4]]
assert make_constant_matrix(0,2,2.4) == [[],[]]
assert make_constant_matrix(2,0,2.4) == []
assert make_constant_matrix(0,0,2.4) == []
```

##Matriz Random

```python
import random
def make_random_matrix(rows: int, cols: int)->list[list[float]]:
  """
  Esta función recibe el tamaño de la matriz que debe de crear (filas y columnas)
  y en cada posición mete un valor constante (value)
  """
  matrix = []
  for i in range(0,cols):
    empty_row = []
    for j in range(0,rows):
      empty_row.append(random.random())

    matrix.append(empty_row)
  return matrix
```

```python
make_random_matrix(2,2)
```

## 2- Determinar indices, de vecinos, para una posicion (x,y)

```python
def neighbors_list(position: list[int], matrix: list[list[float]])->list[int]:
  """
  Recibe una posición en una lista y una matriz.
  Devuelve una lista de listas con los vecinos con la posicion incluida.
  1-Creo un caso general.
  2-Considero los casos especiales.
  """
  neighbors_list = []
  #Meto la posición a la lista de vecinos:
  neighbors_list.append(position)

  #Descompongo la posición en fila y columna:
  row, column = position

  #Vecinos = [[arriba],[abajo],[izquierda],[derecha]]
  #Para un caso general:
  neighbors = [[-1, 0],[1, 0],[0, -1],[0, 1]] # + la posición

  #Iteramos por vecinos
  for neighbor in neighbors:
    new_row = row + neighbor[0]
    new_column = column + neighbor[1]

  #Verificamos que los valores esten dentro de la matriz
    if 0 <= new_row < len(matrix) and 0 <= new_column < len(matrix[0]):
      neighbors_list.append([new_row, new_column])

  return neighbors_list
```

```python
test_matrix = [[1.1, 2.1 ,3.4],[4.2, 5.1, 6.1],[7.1, 8.4, 9.1]]
```

```python
assert neighbors_list([0,0],test_matrix) == [[0, 0], [1, 0], [0, 1]] #Caso Esquina
assert neighbors_list([1,1],test_matrix) == [[1, 1], [0, 1], [2, 1], [1, 0], [1, 2]] #Caso general
assert neighbors_list([2,1],test_matrix) == [[2, 1], [1, 1], [2, 0], [2, 2]] #Caso borde
```

## 3- Sacar el valor de los indices de vecinos

```python
def list_of_values(positions: list[list[int]], matrix: list[list[float]])-> list[int]:
  """
  Recibe una lista de listas con los vecinos y la posicion, y una matriz.
  Devuelve el valor de la posicion y de los vecinos.
  """
  neighbor_values = []
  for pos in positions:
    row = pos[0]
    column = pos[1]

    neighbor_values.append(matrix[row][column])
  return neighbor_values
```

##4- Promedio de valores de vecinos

```python
def average_of_values(list_of_values: list[int])-> float:
  """
  Función que recibe los valores de los vecinos.
  Devuelve el promedio de los valores.
  """
  average = 0
  for element in list_of_values:
    average += element

  return average / len(list_of_values)
```

```python
average_of_values([5, 2, 8, 4, 6])
```

## 5- Average_matrix: Matriz principal que llama a las otras funciones

```python
def average_matrix(matrix:list[list[float]])->list[list[float]]:
  """
  Recibe una matriz de numeros. Devuelve otra matriz, del mismo tamaño,
  con cada punto convertido en el promedio suyo y de sus vecinos.
  Calcuka correctamente los vecinos de los casos frontera (esquinas y bordes)
  """
  new_matrix = []
  for i, column in enumerate(matrix):
    row = []
    for j, value in enumerate(column):
      # Detectar los vecinos:
      neighbours = neighbors_list([i,j], matrix)

      # Obtengo los valores de los vecinos:
      values = list_of_values(neighbours, matrix)

      # Saco el promedio:
      avg = average_of_values(values)

      # Añado cada elementos con el promedio de sus vecinos en la fila:
      row.append(avg)

    #Añado cada fila a la nueva matriz:
    new_matrix.append(row)
  return new_matrix
```
